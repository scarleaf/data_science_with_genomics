```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# 데이터 경로 설정해 주기

```python
train = "../data/house-price/train.csv"
test = "../data/house-price/test.csv"
```

# CSV파일을 Pandas DataFrame으로 만들기

```python
dfTrain = pd.read_csv(train)
dfTest = pd.read_csv(test)
```

# 데이터 feature 확인

```python
dfData = dfTrain
dfData.head()
```

# 데이터 훑어보기

```python
dfData.info()
```

# 데이터 분리 (판매 가격 - 특징들)

```python
dfSalePrice = dfData.SalePrice
dfData.drop(["Id", "SalePrice"], axis=1, inplace=True)
```

# 판매 가격 분포 확인

```python
dfSalePrice.plot.hist(bins=100)
plt.show()
```

```python
sns.distplot(dfSalePrice)
plt.show()
```

## 판매가격 자연로그 스케일 변환

```python
np.log1p(dfSalePrice).to_frame().plot.hist(bins=100)
plt.show()
```

```python
np.log1p(dfSalePrice).plot.hist(bins=100)
plt.show()
```

```python
sns.distplot(np.log1p(dfSalePrice))
plt.show()
```

# 결측치(NaN, . 등)에 대한 정보 확인

```python
na_all = (dfData.isnull().sum() / len(dfData)) * 100
```

```python
na_all_sort = na_all.drop(NA_all[NA_all==0].index).sort_values(ascending=False)
```

```python
dfMissing = pd.DataFrame({"Percentage":na_all_sort})
dfMissing
```

# 결측치에 대한 barplot 그려보기

## Matplotlib

```python
dfMissing.plot.bar(figsize=(10,10))
plt.xticks(rotation="45")
plt.title("Percentage of missing data by feature", fontsize=24)
plt.xlabel("Features", fontsize=16)
plt.ylabel("Percentage of missing values", fontsize=16)
plt.show()
```

## Seaborn

```python
plt.subplots(figsize=(10,10))
sns.barplot(x=dfMissing.index, y=dfMissing.Percentage)
plt.xticks(rotation="45")
plt.title("Percentage of missing data by feature", fontsize=24)
plt.xlabel("Features", fontsize=16)
plt.ylabel("Percentage of missing values", fontsize=16)
plt.show()
```
